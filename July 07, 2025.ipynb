## Numpy

- Python libraries for scientific computing
- Used for applications of machine learning and deep learning
- Used for multi dimensional array operations

## Array

- Array are a collection of elements/values, that can have one or more dimensions.
- An array of one dimension is called a Vector while having two dimensions is called a matrix.
- Numpy arrays are called ndarray or N-dimensional arrays and they store elements of the same type and size.
- It is known for its high performance and provide efficient storage and data operations as arrays grow in size

## Python Lists vs Numpy Arrays

- Numpy arrays that contain only homogeneous element.
- This makes it more efficient at storing and manipulating the array.
