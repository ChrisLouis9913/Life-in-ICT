### Code

```python
import sklearn
from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression
```

```python
X,y=load_iris(return_X_y=True)
lr_clf=LogisticRegression(random_state=0,max_iter=1000)
lr_clf.fit(X,y)
```

```python
print(lr_clf.predict(X[:2, :]))
print(lr_clf.predict_proba(X[:2, :]))
print(lr_clf.score(X,y))
```

```python
lr_clf.classes_
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score
```

```python
X = np.array([[1, 1], [1, 2], [2, 2], [2, 3]])
# y = 1* x_0 + 2 * x_1 + 3
y = np.dot(X, np.array([1, 2])) + 3
reg =LinearRegression()
reg.fit(X,y)
```

```python
print(reg.coef_)
print(reg.intercept_)
print(reg.predict(np.array([[3, 5]])))
```

```python
y_true = y
y_pred=reg.predict(X)
print(mean_squared_error(y,y_pred))
```

## Plotting Techniques

### Line Plot

- **Line plot**

```python
import matplotlib.pyplot as plt
    
x = [0, 1, 2, 3, 4, 5]
y = [0, 3, 2, 22, 22, 10]
plt.plot(x, y)
plt.xlabel("days")
plt.ylabel("counts")
plt.show()
```

- **Line plot Axis controls**

```python
import numpy as np
x = np.arange(0, 400)
y = np.cos(x)
plt.plot(x, y)
plt.xlim(300, 340)
plt.show()
```

- **Multi-line line plots**

```python
x = [0, 1, 2, 3, 4, 5]
y1 = [0, 3, 2, 22, 22, 10]
y2 = [1, 7, 5, 4, 3, 8]
plt.plot(x, y1, c="g", label="location1")
plt.plot(x, y2, c="r", label="location2")
plt.xlabel("days")
plt.ylabel("counts")
plt.legend()
plt.show()
```

### Bar Charts

```python
x = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
item_1_sales = [1, 2, 3, 4, 5]
plt.bar(x, item_1_sales, label="item1")
plt.xlabel("Days")
plt.ylabel("Counts")
plt.savefig("bar_chart.png", dpi=300)
plt.show()
```

* Multi-bar plot

```python
x = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday"]
item_1_sales = [1, 2, 3, 4, 5]
item_2_sales = [3, 4, 5, 6, 5]
plt.bar(x, item_1_sales, color="r", label="item1")
plt.bar(x, item_2_sales, color="b", alpha=0.2,label="item2")
plt.xlabel("Days")
plt.ylabel("Counts")
plt.legend()
plt.savefig("bar_chart.png", dpi=300)
plt.show()
```

### Pie Charts

```python
countries = ["India", "USA", "UK", "UAE"]
population = [123, 20, 30, 40]
plt.pie(population, labels=countries, autopct="%1.2f%%")
plt.show()
```

### Box Plot

```python
a = np.random.normal(170, 1, 20000)
q1 = np.quantile(a, 0.25)
q3 = np.quantile(a, 0.75)
iqr = q3-q1
lower_bound = q1-(1.5*iqr)
upper_bound = q3+(1.5*iqr)
sns.boxplot(a)
```
