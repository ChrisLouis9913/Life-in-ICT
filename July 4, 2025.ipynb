- Inheritance
    
    Inheritance is a fundamental concept in object oriented programming (OOP) where a class (known as the child class or subclass) can inherit attributes and methods from another class (known as the parent class or superclass). This allows the child class to reuse and extend the functionality defined in the parent class.
    
    - Parent Class: The parent class is a class the defines attributes and methods that can be inherited by other classes.
    - Child Class: The child class is a class the inherits attributes and methods from a parent class.
    - Encapsulation: It is the bundling of data (attributes) and methods that operate on the data into a single unit or class.

```python
class Animal:
    def __init__(self, name, age, sound):
        self.name = name
        self.age = age
        self.sound = sound
    
    def make_sound(self):
        print(self.sound)
```

```python
class Dog(Animal):
    def __init__(self, name, age):
        super().__init__(name, age, "Woof!")
    
    def action(self):
        print("Moves tails")
```

```python
class Cat(Animal):
    def __init__(self, name, age):
        super().__init__(name, age, "Meow!")
    
    def action(self):
        print("Moves ears")
```

```python
my_dog = Dog("Buddy", 3)
```

```python
my_dog.make_sound()
my_dog.action()
```

```python
my_cat = Cat("Whiskers", 2)
my_cat.make_sound()
my_cat.action()
```

- Abstract Class

Example 1: Dog

```python
from abc import ABC, abstractmethod
    
class Dog(ABC):  # Abstract Class
    def __init__(self, name):
        self.name = name
    
    @abstractmethod
    def sound(self):  # Abstract Method
        pass
    
    def display_name(self):  # Concrete Method
        print(f"Dog's Name: {self.name}")
```

```python
class Labrador(Dog):
  def sound(self):
    print("Labrador Woof!")
        
class Beagle(Dog):
  def sound(self):
    print("Beagle Bark!")
```

```python
dog1 = Labrador("Buddy")
dog2 =  Beagle("Charlie")
```

```python
dog1.display_name()  # Calls concrete method
dog1.sound()  # Calls implemented abstract method
```

```python
dog2.display_name()
dog2.sound()
```

Example 2: Student

```python
class Student:
    def set_name(self, name, scol, id):
        self.name=name
        self.scol=scol
        self.id=id
            
    def show_name(self):
        print("Name is", self.name)
        print("Shirt color is", self.scol)
        print("ID is", self.id)
        
s=Student()
s.set_name("Louis", "Blue", 5955)
        
s.show_name()
```

```python
class Student1:
    def __init__(self, name, course, id):
        self.name = name
        self.course = course
        self.id = id
    def show_details(self):
        print("Name:", self.name, "Course:", self.course, "Id:", self.id)

s1 = Student1("Chris", "Python", 1)
s2 = Student1("Louis", "Java", 2)
s1.show_details()
s2.show_details()
```

- Polymorphism
    
    Polymorphism means you can write code that works with objects of various types, even if you don't know their exact type at the time you write the code.
    
    - Public: no underscore Eg. self.name
    - Protected: _ single underscore Eg. self._name
    - Private: __ double underscore Eg. self.__name

Example

```python
class mammooty: #parent
    def act(act): #Attributes/Character
        print("superstar")
class dq(mammooty): #child attributes/character
    def dance(dance):
        print("Dance")
    #Creating an instance of a class
son=dq() #child
son.act() #Output: "superstar"
son.dance() #Output: "Dance"
```

```python
class Animal:
    def speak(self):
        print("Animal Speaks")
    
    #Dog inherits from Animal
class Dog(Animal):
    def bark(self):
        print("Dog barks")
    
    #Bulldog inherits from Dog
class Bulldog(Dog):
    def guard(self):
        print("Bulldog Guards")
    #Creating an instance of Bulldog class
Bulldog = Bulldog()
Bulldog.speak()     #Inherited from Animal
Bulldog.bark()      #Inherited from Dog
Bulldog.guard()     #Defined in Bulldog
```
